```python
category_data = {
"Beverages & Milk": ["Cocoa Beverages", "Everyday Tea", "Coffee", "Herbal Teas", "Milk"],
    "Breakfast & Cereals": ["Oats & Instant Cereals", "Sugar, Honey & Sweeteners", "Butter, Cheese & Other Spreads"],
    "Cooking Paste, Oil & Spices": ["Tomato Paste", "Cooking Oils", "Salt & Seasoning Cubes", "Herbs & Spices"],
    "Foodstuff": ["Grains & Rice", "Pasta & Noodles", "Poundo, Wheat & Semolina", "Canned Foods"],
    "Snacks & Confectioneries": ["Biscuits, Chin Chin & Cookies", "Nuts & Seeds", "Chocolates & Sweets", "Dry Fruits"],
    "Baking Ingredients": ["Flour & Baking Powder", "Baking Tools & Accessories"],
    "Alcoholic Drinks": ["Beer", "Liquers & Creams", "Cognac & Spirits", "Wines & Champagne"],
    "Non-Alcoholic Drinks": ["Fizzy Drinks & Malt", "Energy Drinks", "Wines", "Fruit Juices & Yoghurt", "Water"],
    "Baby & Kids": ["Diapering", "Baby & Toddler Health", "Daily Care", "Feeding & Nursing", "Toys & Gears", "School Bag"],
    "Detergent & Laundry Supplies": ["Bar Soaps & Detergents", "Bathroom & Toilet Cleaners", "Fabric Softeners", "Dish Washers", "Glass Cleaner", "Disinfectant & Sprays"],
    "Home Care & Household Supplies": ["Paper Towels & Serviettes", "Foil Paper & Cling Film", "Pests & Insect Control", "Lighters and Match Box", "Air Fresheners"],
    "Beauty & Personal Care": ["Skin Care", "Oral Care", "Hair Care", "Fragrances", "Feminine Care", "Men's Grooming", "Make-up", "Male Shoe", "Female Shoe"]
}
```

```python
def categorize_product(product_name, manufacturer=None):
 
  tokens = product_name.lower().split()
  # Check for specific cases first
  if "coca cola" in manufacturer.lower() and any(token in product_name.lower() for token in ["fizzy", "soda", "cola"]):
    return "Fizzy Drinks & Malt"
  elif any(word in product_name.lower() for word in ["grounded", "soup"]):
    return "Cooking Paste, Oil & Spices"
  elif "iyan" in product_name.lower():
    return "Poundo, Wheat & Semolina"
  elif any(word in product_name.lower() for word in ["liqueur", "rum"]):
    return "Liquers & Creams"

  # General classification based on category_data
  for category, product_types in category_data.items():
    for product_type in product_types:
      if any(token in product_type.lower() for token in tokens):
        return product_type
  return "Unknown Product Type"

```

```python
df['Product Category'] = df.apply(lambda row: categorize_product(row['Product Name'], row['Manufacturer Name'] if 'Manufacturer Name' in row else None), axis=1)

df.head(12)
```

```python
import re

def convert_variant_format(variant):
    # Normalize input by replacing '×' with 'x' and removing spaces around 'x'
    variant = re.sub(r'\s*[xX]\s*', 'x', variant.replace('×', 'x'))
    
    # Pattern to match the format "number g x number"
    pattern1 = re.compile(r'(\d+)\s*g\s*x\s*(\d+)', re.IGNORECASE)
    pattern2 = re.compile(r'(\d+)\s*x\s*(\d+)\s*g', re.IGNORECASE)
    pattern3 = re.compile(r'(\d+)x(\d+)g', re.IGNORECASE)
    pattern4 = re.compile(r'(\d+)X(\d+)ml', re.IGNORECASE)
    pattern5 = re.compile(r'(\d+)\s*cl\s*x\s*(\d+)', re.IGNORECASE)
    

    # Try to match each pattern and convert to "sizeG x number"
    match1 = pattern1.match(variant)
    if match1:
        weight, count = match1.groups()
        return f"{weight.upper()}G x {count}"
    
    match2 = pattern2.match(variant)
    if match2:
        count, weight = match2.groups()
        return f"{weight.upper()}G x {count}"
    
    match3 = pattern3.match(variant)
    if match3:
        count, size = match3.groups()
        return f"{size.upper()}G x {count}"
    
    match4 = pattern4.match(variant)
    if match4:
        count, size = match4.groups()
        return f"{size.upper()}ml x {count}"
    
    match5 = pattern5.match(variant)
    if match5:
        size, count = match5.groups()
        return f"{size.upper()}CL x {count}"

    
    # If no patterns match, return the original variant
    return variant


print(convert_variant_format("800g×6"))     # Should print "800G x 6"
print(convert_variant_format("6 x 280g"))   # Should print "280G x 6"
print(convert_variant_format("6×280g"))     # Should print "280G x 6"
print(convert_variant_format("6x280g"))     # Should print "280G x 6"
print(convert_variant_format("1X330ml"))   # Should print "330ML x 12"
print(convert_variant_format("75CL X 6"))     # Should print "75CL x 6"
print(convert_variant_format("InvalidFormat"))  # Should print "InvalidFormat"
```

```python
df['Variant'] = df['Variant'].apply(convert_variant_format)

df.head(13)
```

```python
# Define functions to extract size and amount
def extract_size(weight_str):
  try:
    # Use regular expression to match a number followed by "kg", "G", or "ml"
    match = re.search(r"(\d+\.?\d*)(kg|G|ml)", weight_str)
    if match:
      value = float(match.group(1)) 
      unit = match.group(2)
      if unit == "kg":
        return value * 1000
      elif unit == "G":
        return value
      elif unit == "ml":
        return value * 1
    return None 
  except:
    return None

def extract_amount(weight_str):
  try:
    amount_start = weight_str.find("x")
    if amount_start == -1:
      amount_start = weight_str.find("×")
    if amount_start == -1:
      return None
    return int(weight_str[amount_start + 1:])
  except:
    return None 


Size = df["Variant"].apply(extract_size)
Amount = df["Variant"].apply(extract_amount)
df["Weight"] = round(((Size * Amount)/1000) + 1)

df.head(12)
```

```python
# Convert Product Name to sentence case
df["Product Name"] = df["Product Name"].str.title()

df['Variant Type'] = "Size"
# Display the updated DataFrame
df.head(12)
```
